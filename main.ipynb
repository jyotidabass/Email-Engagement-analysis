```python

def pkl_to_json(input_file, output_file):
    # Read the data from the pickle file (assuming it's a list)
    data_list = pd.read_pickle(input_file)

    # Convert the list to a DataFrame
    df = pd.DataFrame(data_list)

    # Save the DataFrame to a JSON file
    df.to_json(output_file, orient='records', lines=True)

if __name__ == "__main__":
    # Replace 'input.pkl' and 'output.json' with your actual file names
    input_file = 'email_campaigns.pkl'
    output_file = 'output.json'

    pkl_to_json(input_file, output_file)
```

```python
def show_json_file(json_file):
    with open(json_file, 'r') as file:
        for line in file:
            try:
                data = json.loads(line)
                print(json.dumps(data, indent=2))
            except json.JSONDecodeError as e:
                print(f"Error decoding JSON: {e}")

if __name__ == "__main__":
    # Replace 'output.json' with your actual JSON file name
    json_file = 'output.json'

    show_json_file(json_file)
```

```python
def visualize_json_file(json_file):
    with open(json_file, 'r') as file:
        data_list = [json.loads(line) for line in file]

    # Assuming each JSON object has a common structure, adjust the keys accordingly
    keys_to_visualize = ['subject', 'body', 'opened', 'meeting link clicked', 'responded']

    # Extract data from each JSON object and create a DataFrame
    flattened_data = []
    for item in data_list:
        flattened_data.append({key: item.get(key) for key in keys_to_visualize})

    df = pd.DataFrame(flattened_data)

    # Count occurrences for each key
    counts = {key: df[key].value_counts() for key in keys_to_visualize}

    # Plot counts for each key
    for key, count in counts.items():
        count.plot(kind='bar', stacked=True)
        plt.xlabel(key)
        plt.ylabel('Count')
        plt.title(f'Visualization of {key} in JSON Data')
        plt.show()

if __name__ == "__main__":
    # Replace 'output.json' with your actual JSON file name
    json_file = 'output.json'

    visualize_json_file(json_file)
```

# We first Perform EDA in 'example1' as a demonstration of how we can derive a valuable insight.

```python
egs = df['example1'].tolist()
```

```python
egs
```

```python
opened_subjects = []

for subs in egs:
  if subs != None:
    for mp in subs:
      try:
        if mp['opened'] == True:
          opened_subjects.append(mp['subject'])
      except:
        print(mp)
    # print(subs)
```

```python
opened_subjects.sort()
opened_subjects
```

```python
# Download NLTK stopwords
nltk.download('stopwords')

# Combine all the email subjects into a single string

combined_text = ' '.join(opened_subjects)

# Tokenize the text
tokens = nltk.word_tokenize(combined_text)

# Remove stopwords and non-alphabetic words
stop_words = set(stopwords.words('english'))
filtered_tokens = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stop_words]

# Create a frequency distribution of words
freq_dist = nltk.FreqDist(filtered_tokens)

# Print the top words
top_words = freq_dist.most_common(10)
print("Top Words:", top_words)

# Create a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(freq_dist)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Email Subjects')
plt.show()
```

These are the top words used to get the ['Open'= True] with respect to the 'subject' in the above dataset. The exact similar process could be done for getting an analysis of what prompted the users to book a meeting (i.e. 'meeting link clicked': True) or (respond i.e. 'responded': True) by subsituting these for 'Open' in the above program and 'body' instead of 'subject' going by the logic that users tend to respond or click the meeting link only if they are interested.
